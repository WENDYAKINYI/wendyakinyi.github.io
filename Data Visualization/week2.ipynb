```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import timedelta
import yfinance as yf
```

```python
# Assignment 2, Part 1: Create a list of the atomic weights of the first six elements of the periodic table, each rounded to the nearest integer. Provide two pie charts as follows: (1) each slice annotated with a percentage of the whole, and (2) each slice annotated with its atomic weight.  Explode a different element with each chart.
elements = ["H", "He", "Li", "Be", "B", "C"]
rounded_atomic_weights = [1, 4, 7, 9, 11, 12]
print(elements, rounded_atomic_weights)

# Explode a different element in each chart
explode_percentage = (0, 0.1, 0, 0, 0, 0)  # Explode 'He' in the first chart
explode_weight = (0, 0, 0.1, 0, 0, 0)  # Explode 'Li' in the second chart


# Pie chart with percentage annotations
plt.figure(figsize=(10, 5))
plt.pie(rounded_atomic_weights, labels=elements, autopct='%1.1f%%', explode=explode_percentage)
plt.title('Atomic Weights of the First Six Elements with Percentage Annotations')
plt.axis('equal')
plt.show()

# Pie chart with atomic weight annotations
plt.figure(figsize=(10, 5))
plt.pie(rounded_atomic_weights, labels=elements, autopct=lambda p: '{:.0f}'.format(
    p * sum(rounded_atomic_weights) / 100), explode=explode_weight)
plt.title('Atomic Weights of the First Six Elements with Atomic Weight Annotations')
plt.axis('equal')
plt.show()
```

```python
# Assignment 2, Part 2: Read into a DataFrame the file py_ide2.csv and provide both a horizontal bar chart and a vertical bar chart, complete with all labels. Be sure to rotate the IDE names so that they are readable.
# Read the CSV file into a DataFrame
df = pd.read_csv('py_ide2.csv')
# Horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(df['IDE'], df['Adoption'], color='skyblue')
plt.xlabel('Adoption')
plt.ylabel('IDE')
plt.title('Horizontal Bar Chart of Python IDEs')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Vertical bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['IDE'], df['Adoption'], color='lightgreen')
plt.xlabel('IDE')
plt.xticks(rotation=45)
plt.ylabel('Adoption')
plt.title('Vertical Bar Chart of Python IDEs')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Assignment 2, Part 3: Construct a list of eight strings that represent days evenly spread out. Drawing from the random uniform distribution, make an array of eight floats ranging from 100 to 200 in value.  Establish a DataFrame from that list and that array, convert the dates to pandas datetime objects and set them to the index.  Make two charts in the same window or canvas as follows: (1) line plot of the values vs. dates, and (2) bar chart of the same.
# List of dates evenly spread over 8 days
date_list = [pd.Timestamp.today() + timedelta(days=i) for i in range(8)]
print(date_list)

# Generate random values between 100 and 200
value_list = np.random.uniform(low=100, high=200, size=8)
print(value_list)

# Create DataFrame with dates as index and values as column
df = pd.DataFrame(
    {'Values': value_list, 'Dates': date_list})
print(df)

# Convert index to datetime format
df.set_index('Dates', inplace=True)

# Create a figure for plotting
plt.figure(figsize=(10, 6))

# Line plot
plt.subplot(121)
plt.plot(df.index, df['Values'], marker='o', linestyle='-', color='b')
plt.xlabel('Dates')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.title('Line Plot - Values vs. Dates')

# Bar chart
plt.subplot(122) 
plt.bar(df.index, df['Values'],  color='g')
plt.xlabel('Dates')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.title('Bar Chart - Values vs. Dates')
plt.tight_layout()
plt.show()
```

```python
# Assignment 2, Part 4: Pull from Yahoo Finance the closing prices and volumes of the stock of your choice over the trading days of one month and plot the prices and volumes on a canvas in two separate panels, one above the other, with the dates aligned.


# Define the stock symbol of the stock
stock_symbol = 'NVDA'  # NVIDIA Corporation Stock

# Define the date range over the trading days of one month (March 2024)
start_date = '2024-03-01'
end_date = '2024-03-31'

# Pull data from Yahoo Finance
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

# Extract closing prices and volumes
closing_prices = stock_data["Close"]
volumes = stock_data["Volume"]

# Create a figure and two subplots
fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot closing prices
ax0.plot(stock_data.index, closing_prices, color='b')
ax0.set_title('Closing Prices of NVDA Stock')
ax0.set_ylabel('Price')

# Plot volumes
ax1.bar(stock_data.index, volumes, color='r')
ax1.set_title('Volumes of NVDA Stock')
ax1.set_ylabel('Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
